```python
import pandas as pd
import numpy as np
import seaborn as sns
import os
import warnings
from scipy import stats
from scipy.stats import normaltest, boxcox, yeojohnson
import matplotlib.pyplot as plt
```

```python
data = pd.read_csv(
    "../data/external/public_train_features.csv", index_col="row_id"
)
```

```python
data.select_dtypes(include=["float"]).columns
```

```python
data["diff_time_gate_lag_1"].max()
```

```python
plt.figure(figsize=(16, 8))
plt.subplot(1, 2, 1)
stats.probplot(data["diff_time_to_sec"], dist="norm", plot=plt)
plt.title("diff_time_to_sec распределение")
plt.subplot(1, 2, 2)
stats.probplot(data["second_lag_2"], dist="norm", plot=plt)
plt.title("second_lag_2 распределение")
```

```python
data["diff_time_gate_lag_1_boxcox"], param_price_boxcox = boxcox(
    data["diff_time_gate_lag_1"]
)
print(f"Параметр Бокса-Кокса price: {param_price_boxcox}")
plt.figure(figsize=(16, 8))
plt.subplot(1, 2, 1)
stats.probplot(data["diff_time_gate_lag_1"], dist="norm", plot=plt)
plt.title("diff_time_gate_lag_1 распределение")
plt.subplot(1, 2, 2)
stats.probplot(data["diff_time_gate_lag_1_boxcox"], dist="norm", plot=plt)
plt.title("diff_time_gate_lag_1_boxcox распределение")
```
